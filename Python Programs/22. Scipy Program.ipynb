```python
# imporing the libraties
from sympy import *
from sympy.parsing.sympy_parser import parse_expr
# defining the function
def definite_integral(expr, symbol, x_range):
  # parsing the expression
  expr = parse_expr(expr)
  # getting the symbol
  x = symbols(symbol)
  # finding the result
  integration = integrate(expr, (x, x_range[0], x_range[1]))
  # returing the result
  return integration
# use if you want to get inputs from user
# expr = input("Enter the expression: ")
# symbol = input("Enter the symbol: ")
# t = input("Enter the range: ")
# x_range = tuple(int(x) for x in t.split(','))
# print(definite_integral(expr,symbol,x_range))

# use if you want to give the values predefined
print(definite_integral('x**2', 'x', (-1,1)))
```
