```python
// importing the libraries
import java.util.ArrayList;
import java.util.Arrays;
import java.util.Random;
// initializing the main class
public class Main {
        // initializing the main method
        private static void combinationsOfSum(int[] array, ArrayList<Integer> list) {
                if(list.isEmpty())
                        return;
                print(array,list);
                // getting index of last element of combination
                int top=list.get(list.size()-1);
                // Removing last element of list 
                list.remove(list.size()-1);
                // checking for all combination are over or not
                if(top==4 && list.isEmpty())
                        return;
                // iterate again 
                if(top<4) {
                        list.add(top);
                        list.add(top+1);
                }
                else{
                        int last=list.get(list.size()-1); 
                        list.remove(list.size()-1);
                        list.add(last+1);
                }               
                // calling the function
                combinationsOfSum(array, list);
        }
        // displaying the comibations
        private static void print(int[] array, ArrayList<Integer> list) {
                String s="";
                long sum=0;
                for(Integer i:list) {
                        s+=array[i]+" + ";
                        sum+=array[i];
                }
                s=s.substring(0,s.length()-2)+"= "+sum;
                System.out.println(s);
        }
        // initializing the main method
        public static void main(String[] args) {
                int array[]=new int[5];
                
                // Random class
                Random rd=new Random();
                // getting the random values 
                for(int i=0;i<5;i++)
                        array[i]=rd.nextInt(100);
                // displaying the array
                System.out.print("Given array = {");
                for(int i=0;i<5;i++)
                    System.out.print(array[i]+" ");
                System.out.print("}, all summed combination are: \n");
                // sortting array
                Arrays.sort(array);
                // tracking of current combination 
                ArrayList<Integer> list=new ArrayList<Integer>();
                list.add(0);
                // calling the function
                combinationsOfSum(array,list);
        }
        
}
```
