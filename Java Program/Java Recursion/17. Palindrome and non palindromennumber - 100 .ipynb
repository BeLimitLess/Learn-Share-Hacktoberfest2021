```python
// importing the libraries
import java.util.*;
// initializing the main class
public class Main{
    // initializing the main method
        public static void main(String[] args){{
            int number=12,number2,rangeGiven=0,i;
            int[] prime=new int[100];
            while(rangeGiven<100){
                if(!check(number,prime,rangeGiven))
                    if(checkPrime(number)){
                        number2=reverse(number);
                        if(number2!=number)
                            if(!check(number2,prime,rangeGiven))
                                if(checkPrime(number2))
                                    rangeGiven=insert(number,number2,prime,rangeGiven);
                }
                number++;
            }
            // function calling
            sort(prime,rangeGiven);
            // displaying the result
            System.out.println("The first 100 EMIRP numbers are:");
            for(i=0;i<100;i++){
                if(i%10==0)
                    System.out.println();
                    System.out.printf("%7d",prime[i]);
            }
        }
}
// method for sorting
public static void sort(int a[],int n){
    int i,j,t;
    for(i=0;i<n-1;i++)
        for(j=i;j<n;j++)
            if(a[i]>a[j]){
                t=a[i];
                a[i]=a[j];
                a[j]=t;
            }
    }
    public static boolean check(int n,int a[],int c){
        int i;
        for(i=0;i<c;i++)
            if(a[i]==n)
            return true;
        return false;
    }
    public static int insert(int n,int m,int a[],int i){
        a[i++]=n;
        a[i++]=m;
        return i;
    }
    // function to reverse the number
    public static int reverse(int n){
        int numberNew=0;
        while(n>0){
            numberNew=numberNew*10+n%10;
            n=n/10;
        }
        return numberNew;
    }
    // checking the number prime or not
    public static boolean checkPrime(int n){
        int i;
        for(i=2;i<n;i++)
            if(n%i==0)
            return false;
        return true;
    }
}
```
