```python
import java.util.*;
public class Main{
        public static void main(String[] args) {
                Scanner input = new Scanner(System.in);
                String str = input.next(); // taking input from user
                char letter1 = str.charAt(0); // taking first char from input
                char letter2 = ' ';
                if (str.length() >= 2){ 
                        letter2 = str.charAt(1); // taking second char from input
                }
                double gradeValue = 0;
                if (letter1 == 'A'){
                        if (letter2 == ' ')
                                gradeValue=4.00;
                        else if (letter2 == '+')
                                gradeValue=4.25;
                        else if (letter2 == '-')
                                gradeValue=3.75;
                }
                else if (letter1 == 'B'){
                        if(letter2 == ' ')
                                gradeValue=3.00;
                        else if(letter2 == '+')
                                gradeValue=3.25;
                        else if(letter2 == '-')
                                gradeValue=2.75;
                }

                                
                else if(letter1 =='C'){
                        if(letter2 == ' ')
                                gradeValue=2.00;
                        else if(letter2 == '+')
                                gradeValue=2.25;
                        else if(letter2 == '-')
                                gradeValue=1.75;
                }
                                
                else if(letter1 == 'D'){
                        if(letter2 == ' ')
                                gradeValue=1.00;
                        if(letter2 == '+')
                                gradeValue=1.25;
                        if (letter2 ==  '-')
                                gradeValue=0.75;
                }

                                
                else if(letter1 == 'F'){
                        if(letter2 == ' ')
                                gradeValue=0.00;
                        if (letter2 == '+')
                                gradeValue=0.25;
                        if (letter2 ==  '-')
                                gradeValue=0.00; // i am taking F- as 0 , if not than change to your wish.
                }

                else{
                        gradeValue = 0.0;
                        System.out.println("Invalid Grade");
                }
                System.out.println("Grade is: " + gradeValue);
        }
}
```
