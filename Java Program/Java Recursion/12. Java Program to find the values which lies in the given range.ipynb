```python
// importing the libaries
import java.util.Scanner;
// declaring the main class
public class Main{
    // declaring teh main method
    public static void main(String[] args) {
        // taking the scanner class
        Scanner sc = new Scanner(System.in);
        // local variable to store the size
        int size = sc.nextInt();
        // declaring array 
        int[] userListArray = new int[size];
        // getting the elements of the array
        for(int i=0;i<size;i++){
            // appending the elements to the array
            userListArray[i] = sc.nextInt();
        }
        // boundary value is set to 50
        int boundsValue = 50;
        // iterating over the elements
        for(int i=0;i<size;i++){
            // checking the values that lies in between 
            // 0 and 50 bounds
            if(userListArray[i]<=boundsValue){
                // displaying the result
                System.out.print(userListArray[i]+",");
            }
        }
    }
}
```
