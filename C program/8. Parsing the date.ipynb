```python
// importing the libraries
#include <stdio.h>
#include<stdlib.h>
#include<ctype.h>
#include <string.h>
#define MAX 100
// converting the month to numbers
// in series
int monthConversion(char *monthName) {
   int monthNumber;
   if (strcmp(monthName, "January") == 0) {
       monthNumber = 1;
   }
   else if (strcmp(monthName, "February") == 0) {
       monthNumber = 2;
   }
   else if (strcmp(monthName, "March") == 0) {
       monthNumber = 3;
   }
   else if (strcmp(monthName, "April") == 0) {
       monthNumber = 4;
   }
   else if (strcmp(monthName, "May") == 0) {
       monthNumber = 5;
   }
   else if (strcmp(monthName, "June") == 0) {
       monthNumber = 6;
   }
   else if (strcmp(monthName, "July") == 0) {
       monthNumber = 7;
   }
   else if (strcmp(monthName, "August") == 0) {
       monthNumber = 8;
   }
   else if (strcmp(monthName, "September") == 0) {
       monthNumber = 9;
   }
   else if (strcmp(monthName, "October") == 0) {
       monthNumber = 10;
   }
   else if (strcmp(monthName, "November") == 0) {
       monthNumber = 11;
   }
   else if (strcmp(monthName, "December") == 0) {
       monthNumber = 12;
   }
   else {
       monthNumber = 0;
   }
   return monthNumber;
}
// initializing the main function
int main() {
    // local variables
    char input[MAX],dayValue[MAX],yearValue[MAX],data[MAX][MAX];
    int i,k,j,index=0,size=0,day,year;
    char month[MAX];
    char *temp;
    // while loop to iterate till the user enters -1 
    do{
        fgets(input,100,stdin);
        if(strlen(input)==0){
             continue;
        }
        if(strcmp(input,"-1\n")==0){
             break;
        }strcpy(data[size],input);
        size++;
    }while (1);

    for(j=0;j<size;++j){
     strcpy(input,data[j]);
     //checking for / or , break
     temp = strstr(input,"/");
     if(temp!=NULL){
         continue;
     }
     temp = strstr(input,",");
     if(temp==NULL){
         continue;
     }
     if(strlen(input)!=0 && !isalpha(input[0])){
         continue;
     }
    for(i=0;i<strlen(input);++i){
     if(input[i]==' '){
       index=i;
       break;   
     }
    
    }
     k=0;
    // findindg the month
    for(i=0;i<index;++i){
         month[k++]=input[i];
    }
  
    month[k]='\0';

    k=0;
    for(i=index+1;input[i]!=',';++i){
        dayValue[k++]=input[i];
    }
    dayValue[k]='\0';
    // converting to number
    day = atoi(dayValue);
    // finding year
    k=0;
    i++;
    for(;input[i]!='\n';++i){
        yearValue[k++]=input[i];    
    }
    yearValue[k]='\0';
    // converting char to integer
    year =atoi(yearValue);
    printf("%d/%d/%d\n",monthConversion(month),day,year);
    } 
   return 0;
}
```
