```python
 # declaring the game
A = [[0,1,2],[3,4,5],[6,7,8]]             
B = [[0,1,2],[3,4,5],[6,7,8]]
C = [[0,1,2],[3,4,5],[6,7,8]]
# condition to check the constraints of the board
def check_board(board):                
    # checking for 3x3 rows
    for x in range(len(board)):           
        # disabling the row for every true
        disable = True                    
        # checking for columnwise
        for y in range(len(board)):       
            if board[x][y] != "X":        
                disable = False
                continue
        # returing the true for every disable which is true
        if disable == True:              
            return(disable) 
    # checking for 3x3 columns      
    for x in range(len(board)): 
        disable = True
        # checking for rowwise
        for y in range(len(board)): 
            if board[y][x] != "X": 
                disable = False
                continue
        # returing the true for every disable which is true      
        if disable == True: 
            return(disable) 
    # checking for diagonal
    disable = True
    y = 0
    # checking for row and columns wise 
    for x in range(len(board)):  
        if board[x][x] != "X": 
            disable = False
    if disable: 
        return disable 
    disable = True
    # checking for other diagonals
    if disable: 
        for x in range(len(board)): 
            y = len(board) - 1 - x
            if board[x][y] != "X": 
                disable = False
    return disable 
def display():                                                         
    # checking whether the board are disabled or not
    if not check_board(A) and not check_board(B) and not check_board(C): 
        print("A\tB\tC")                                                 
        for i in range(3):
            print(*A[i],end="\t")                                        
            print(*B[i],end="\t")
            print(*C[i])          
    # checking two boards are not disabled
    elif not check_board(A) and not check_board(B):                    
        print("A\tB")
        for i in range(3):
            print(*A[i],end="\t")
            print(*B[i])  
    elif not check_board(A) and not check_board(C):                    
        print("A\tC")
        for i in range(3):
            print(*A[i],end="\t")
            print(*C[i])          
    elif not check_board(B) and not check_board(C):                   
        print("B\tC")
        for i in range(3):
            print(*B[i],end="\t")
            print(*C[i])
    # checking if any board is not visible
    elif not check_board(A):                                        
        print("A")
        for i in range(3):
            print(*A[i])        
    elif not check_board(B):                                       
        print("B")
        for i in range(3):
            print(*B[i])          
    elif not check_board(C):                                       
        print("C")
        for i in range(3):
            print(*C[i])
def player_win():                                                  
    if check_board(A) and check_board(B) and check_board(C):       
        return True
    return False
i = 0                                                            
while(not player_win()):                                         
    display()
    s = ""
    while(1):                                                   
        if i%2==0:                                            
            s = input("Player 1: ")
        else:
            s = input("Player 2: ")
        
        if s[0]>="A" and s[0]<="C" and s[1]>="0" and s[1]<="8" and len(s)==2:  
            board = ""                                           
            x,y = -1,-1                                          
            if s[0]=="A":                                        
                board = A
            elif s[0] == "B":
                board = B
            else:
                board = C
            
            x = int(s[1])//3                                   
            y = int(s[1])%3                                    
            if not check_board(board) and board[x][y] != "X":  
                board[x][y] = "X"
                break
            else:                                              
                print("Invalid move, please input again")
        else:
            print("Invalid move, please input again")
    i+=1

# displaying which player wins
if i%2==0:                                                  
    print("Player 1 wins game")                             
else:
    print("Player 2 wins game")
```
