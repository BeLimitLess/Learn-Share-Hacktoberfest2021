```python
# getting user input
userInput = input()
# converting to list
convertList = list(map(int,userInput.split()))
# popping the value rm list
lastVal = convertList.pop()
# local variable to store count
count = 0
# for loop to iterate over list
for i in convertList:
    # converting and getting the value
    binaryVal = "".join(reversed(bin(i)[2:]))
    try:
        # checking the least significant bit position
        if binaryVal[lastVal] == "1":
            # incrementing the value
           count += 1
    except:
        # pass
        pass
# displaying the result
print(count)
```
