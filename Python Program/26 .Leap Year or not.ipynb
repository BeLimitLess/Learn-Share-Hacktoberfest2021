```python
# getting input from the user
year = int(input())
# checking if the number entered is divisible by 4 or not
if (year % 4) == 0:
  # checking if the number entered is divisible by 100 or not
   if (year % 100) == 0:
     # checking if the number entered is divisible by 400 or not
       if (year % 400) == 0:
           # displaying the result
           print("{0} - leap year".format(year))
       else:
          # displaying the result
           print("{0} - not a leap year".format(year))
   else:
      # displaying the result
       print("{0} - leap year".format(year))
else:
  # displaying the result
   print("{0} - not a leap year".format(year))
```
