```python
# user list
l=[]
# defining the function 
def table_statistics(FileName):
    # opening the file
    f = open(FileName, "r")
    # reading each line in a file
    Lines = f.readlines() 
    # iterate and split the data
    # in file
    for line in Lines:
        l.append(list(line.split('\\n')))
    # sorting the data
    l.sort(key=lambda x: x[2])
    # calculating the average
    avg=0.0
    for i in l:
        avg+=float(i[2])
    # displaying the result
    print("The number of records in the file:",len(l))
    print("The most Expensive dining table model name:",l[len(l)-1][0])
    print("The most Expensive dining table description:",l[len(l)-1][1])
    print("The most Expensive dining table price:",l[len(l)-1][2])
    print("The cheapest dining table model name:",l[0][0])
    print("The cheapest dining table description:",l[0][1])
    print("The cheapest dining table price:",l[0][2])
    print("The average price of all dining table in the file:",avg/len(l))
# calling the function
table_statistics("DiningTable.txt")  
    
```
