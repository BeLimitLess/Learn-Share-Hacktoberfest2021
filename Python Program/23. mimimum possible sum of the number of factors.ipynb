```python
# define function to calculate factors
def factor(b):
    count = 0
    # for loop to iterate over the value
    for i in range(1,b+1):
        if b%i == 0:
            count += 1
    return count
# getting the number of elements 
n = int(input())
x = []
# taking the values into list 
for i in range(n):
  val = int(input())
  x.append(val)
for i in x:
    summ = 0
    if i <= 1:
        print(i)
    else:
      # calculating the result
        for a in range(1,i):
            summ += factor(a)
            if summ >= i:
                print(a)
                break
```
