**Write a simple airline ticket reservation program. The information is maintained on an alphabetized linked list of names. In a simpler version of the program, assume that tickets are reserved for only one flight. In a fuller version, place no limit on the number of flights. Create a linked list of flights with each node including a pointer to a linked list of passengers.**

```python
// including header files
#include<iostream>
#include<string>
using namespace std;
// implementation of Passenger class
class Passenger{
public:
     //Declare lastName and flightName as type of string
     string lastName, flightName;
     // Declare pointer to next node
     Passenger *link;
     // Implementation of parameterized constructor
     Passenger(const string lastName, string flightName, Passenger* pObject = NULL){
          this->lastName = lastName;
          this->flightName = flightName;
          this->link = pObject;
     }
};
// implementating of Ticket class
class Ticket{
private:
     // Declaring header and tail pointer as type of Passenger class
     Passenger *headPointer, *tailPointer;
public:
     // Implementation of parameterized constructor
     Ticket(){
          headPointer = NULL;
          tailPointer = NULL;
     }
     // Implementation of cancel ticket function
     void cancleTicket(){
          // Declare the lastName,flightName as type
          //of string
          string lastName, flightName;
          // Get the passenger name from user
          cout << "Enter the passenger name" << endl;
          cin >> lastName;
          //Get the flight name from user
          cout << "Enter the flight name" << endl;
          cin >> flightName;
          //Creation of an object for Passenger class
          Passenger *passangerObject = headPointer;
          Passenger *passanger2Object = headPointer;
          // check Passenger name and flight name matches with appropriate
          //flight name and last name
          if ((passangerObject->lastName == lastName) && (passangerObject->flightName == flightName)){
              //Set head as head of link node
              headPointer = headPointer->link;
              //Display statement for ticket is cancelled
              cout << "Ticket is cancelled" << endl;
              delete passangerObject;
              return;
          }
          passangerObject = headPointer->link;
          /* Iterate the loop to Delete all nodes of
          from linked list that matches with given apporpriate user data */
          while (passangerObject){
              //compare the last name and flight name
              if ((passangerObject->lastName == lastName) && (passangerObject->flightName == flightName)){
                   passanger2Object->link = passangerObject->link;
                   //delete
                   delete passangerObject;
                   // Display statement
                   cout << "Successful Cancelled the Ticket " << endl;
                   return;
              }else{
                   passanger2Object = passangerObject;
                   passangerObject = passangerObject->link;
              }
          }
          // Display statement
          cout << "No such passengers are exist" << endl;
          return;
     }
     //Implementation of destructor
     ~Ticket(){
          //Declaration of Passenger class objects as passangerClass ,passanger2Class
          Passenger *passangerClass = headPointer;
          Passenger *passanger2Class = headPointer;
          // Iterate the loop and delete the nodes of the given
          //linked list .
          while (passangerClass != NULL){
              passanger2Class = headPointer->link;
              //delete passangerClass
              delete passangerClass;
              passangerClass = passanger2Class;
          }
     }
     //Implementation of reserveTicket function
     void reserveTicket(){
          // Declare the variables lastName, flightName as type of string
          string lastName, flightName;
          //Get the input from user
          cout << "Enter the passenger name" << endl;
          cin >> lastName;
          //Get the input from user
          cout << "Enter the flight name" << endl;
          cin >> flightName;
          //check headPointer is equal to NULL
          if (headPointer == NULL){
              headPointer = new Passenger(lastName, flightName);
              tailPointer = new Passenger(lastName, flightName);
              // Display statement
              cout << "Successful Booking the ticket for passenger" << endl;
              return;
          }
          else{
              //Declare pObject as pointer type object of Passenger class and assign headPointer
              Passenger *pObject = headPointer;
              //Set to link node of list
              while (pObject->link){
                   //check lastName is less than pObject->lastName
                   if(lastName < pObject->lastName){
                        break;
                   }else{
                        pObject = pObject->link;
                   }
              }
              pObject->link = new Passenger(lastName, flightName, pObject->link);
              // Display statement
              cout << "Success" << endl;
              return;
          }
     }
     //Implementation of checkTicket function
     void checkTicket(){
          // Declare the variables
          string lastName;
          //Get the input from user
          cout << "Enter the passenger name" << endl;
          cin >> lastName;
          // Set pointer to linked listâ€™s head node
          Passenger* pObject = headPointer;
          //Iterate the loop
          while (pObject != NULL){
              //check pObject->lastName is equal to lastName
              if (pObject->lastName == lastName){
                   //Display statements the Passenger details
                   cout << "Passenger Name" << endl;
                   cout << pObject->lastName << endl;
                   cout << "Flight Name" << endl;
                   cout << pObject->flightName << endl;
                   return;
              }else{
                   pObject = pObject->link;
              }                 
          }
          //Display statement
          cout << "Success" << endl;
          return;
     }
     //Implementation of displayPassangersList function
     void displayPassangersList(){
          //Set the pointer to headPointer of the given list
          Passenger* pObject = headPointer;
          //Iterate the loop
          while (pObject != NULL){
              //Display statements the Passenger details
              cout << "Passenger Name" << endl;
              cout << pObject->lastName << endl;
              cout << " Flight Name" << endl;
              cout << pObject->flightName << endl;
              //connect to link node of the given linked list
              pObject = pObject->link;
          }
     }
};
int main(){
     //Declaration of tObject for Ticket class
     Ticket tObject;
     //Iterate the loop
     while (true){
          //Display statement for menu
          cout << "Airline Ticket Reservation Program" << endl;
          cout << "1. Book a Ticket" << endl;
          cout << "2. Cancel a Ticket" << endl;
          cout << "3. Check a Ticket" << endl;
          cout << "4. Display the Passengers details" << endl;
          cout << "5. Quit" << endl;
          cout << "Enter your choice" << endl;
          //Declare userChoice as integer type
          int userChoice;
          cin >> userChoice;
          //check with userChoice
          switch (userChoice){
              case 1:
                   //call reserveTicket function
                   tObject.reserveTicket();
                   break;
              case 2:
                   //Ã§all cancleTicket function
                   tObject.cancleTicket();
                   break;
              case 3:
                   //Ã§all checkTicket function
                   tObject.checkTicket();
                   break;
              case 4:
                   //Ã§all displayPassangersList function
                   tObject.displayPassangersList();
                   break;
              case 5:
                   return 0;
              default:
                   //Display statement
                   cout << "Invalid choice .Please try again" << endl;
                   break;
          }
     }
     return 0;
}
```
