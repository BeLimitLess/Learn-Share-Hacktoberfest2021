Sea level Yearly Rise.png

```python
#include <iostream>
#include <iomanip>
using namespace std;

int main() {
  float rise;
  while (1) {
    cout << "What is the water level rise per year: ";
    cin >> rise;
    if (rise >= 0 && rise < 8) {
      break;
    }
  }
  float ans = rise + 0.00;
  cout << "Year Rise\n==============" << endl;
  for (int i = 1; i <= 25; i++) {
    cout << fixed << setprecision(2) << i << "  " << setw(2) << ans << " mm" << endl;
    cout << "-----------" << endl;
    ans += rise;
  }
}
```
